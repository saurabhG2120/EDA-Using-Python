```python

#loading new dataset
df_country=pd.read_excel("Country-Code.xlsx")
```

```python
df_country.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how="left")
```

```python
final_df.headad()
```

```python
final_df.dtypes
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_values=final_df.Country.value_counts().values
```

```python
##creating pie chart for top 3 countries that uses zomato
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
```

obersvation: zomato maximum transaction are in india after that USA and then UK
it shows that major business of zomato happens in india 

```python
final_df.columns
```

```python
#.reset_index()==>to convert the obtained data into data frame
#.rename(columns={0:"rating count"}==> to rename the column

ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"rating count"})
```

```python
ratings
```

 # observations
  1.when rating ranges from 4.5 to 4.9 -->excellent
     2.when rating ranges from 4.0 to 4.4 --> very good 
        3.when rating ranges from 3.5 to 3.9 -->good
          4.whe rating ranges from 2.5 to 3.4 -->average
            5.when rating ranges from 1.8 to 2.4 -->poor
               rest are non rated

```python
ratings.head()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Aggregate rating",y="rating count",data=ratings)
```

```python
#palette is use to map colors 
plt.figure(figsize=(12,6))
sns.barplot(x="Aggregate rating",y="rating count",hue="Rating color",data=ratings,palette=["blue","red","orange","yellow","green","green"])
```

# observations
  1.not rated count is very high
  2. Maximum number of ratings are between 2.5 to 3.4
  
  

```python
sns.countplot(x="Rating color",data=ratings,palette=["blue","red","orange","yellow","green","green"])
```
